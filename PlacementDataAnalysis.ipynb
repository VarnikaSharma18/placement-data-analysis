# Mini Project: Placement Data Analysis



## Objective

The objective of this project is to analyze student placement data and identify how factors like CGPA, skills, and internships affect placement status and salary packages.

### Importing required libraries 

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Loading the Dataset


```python
df = pd.read_csv("placement.csv")
df.head()
```

### Average and spread of salary packages


```python
print("Average Package:", np.mean(df['package_lpa']))
print("Maximum Package:", np.max(df['package_lpa']))
print("Minimum Package:", np.min(df['package_lpa']))
```

### Standard deviation of salary

```python
print("Standard Deviation of Package:", np.std(df['package_lpa']))

```

## 2. Dataset Understanding


To see what are the columns and data types?

```python
df.info()
```

### Average CGPA of placed vs not placed students


```python
print(df.groupby('placement_status')['cgpa'].mean())
```

**Observation:**
Placed students have a higher average CGPA compared to not placed students. This indicates that CGPA plays an important role in placement.

## 3. Data Preprocessing

To check if there are any missing values.

```python
df.isnull().sum()
```

### Dataset Details:
- Total Records: 300
- Total Columns: 14
- Important Variables:
  - CGPA
  - Skills Count
  - Internships
  - Placement Status
  - Package (LPA)

## 4. Filling missing values


```python
df['job_role'] = df['job_role'].fillna("Not Placed")
```

**Action Taken:**
Missing values in the 'job_role' column were replaced with "Not Placed" because students who are not placed do not have any assigned job roles.

## 4. Visualization and Analysis

### (A) Bar Chart


Representing the data of students who are placed vs not placed.

```python
plt.figure()
sns.countplot(x='placement_status', data=df)
plt.title("Placement Status")
plt.xlabel("Placed / Not Placed")
plt.ylabel("Number of Students")
plt.show()
```

**Explanation:**
This bar chart shows the number of students who are placed and not placed. From the graph, we can observe the overall placement rate in the dataset. It helps in understanding how many students successfully got placed.

### (B) Box Plot


Representing relation between CGPA and placement.

```python
plt.figure()
sns.boxplot(x='placement_status', y='cgpa', data=df)
plt.title("CGPA vs Placement Status")
plt.xlabel("Placement Status")
plt.ylabel("CGPA")
plt.show()
```

**Explanation:**
This box plot compares the CGPA distribution of placed and not placed students. It shows that students who are placed generally have higher CGPA values. This indicates that CGPA plays an important role in placement.

### (C) Scatter Plot


Representing the relationship between skills and salary.

```python
plt.figure()
sns.scatterplot(x='skills_count', y='package_lpa', data=df)
plt.title("Skills vs Salary Package")
plt.xlabel("Number of Skills")
plt.ylabel("Package (LPA)")
plt.show()
```

**Explanation:**
This scatter plot shows the relationship between number of skills and salary package. We can observe a positive trend where students with more skills tend to receive higher salary packages. This suggests that skills significantly impact placement outcomes.

### (D) Line Chart


To Show the change in average package with internships.

```python
avg_package = df.groupby('internships')['package_lpa'].mean()

plt.figure()
avg_package.plot(marker='o')
plt.title("Average Package vs Internships")
plt.xlabel("Number of Internships")
plt.ylabel("Average Package (LPA)")
plt.show()
```

**Explanation:**
This line chart shows how the average salary package changes with the number of internships. It indicates that internships have an influence on salary, although the trend is not strictly increasing. Overall, internships contribute positively to placement opportunities.

### (E) Histogram


To Show the distribution of salary packages.

```python
plt.figure()
sns.histplot(df['package_lpa'], bins=10)
plt.title("Salary Package Distribution")
plt.xlabel("Package (LPA)")
plt.ylabel("Frequency")
plt.show()
```

**Explanation:**
This histogram represents the distribution of salary packages among students. Most of the packages are concentrated in the lower to mid range, while very high packages are less frequent. This shows that only a few students receive very high salaries.

## 5. Key Insights

- More skills lead to higher salary packages  
- Internships improve placement opportunities
